```python
x = np.linspace(-20, 20, 1000)


d_1=stats.norm.pdf(x, 0, 1)
d_2=stats.norm.pdf(x, 10, 1)
d_3=stats.t.pdf(x, df=2, loc=0)
d_4=stats.t.pdf(x, df=2, loc=10)

```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,9))

#row, column

axs[0, 0].plot(x, d_1)
axs[0, 0].set_title('Std. Normal w/  mean 0, var.1', fontweight='bold')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('p(x)')
axs[0, 1].plot(x, d_2)
axs[0, 1].set_title('Std. Normal w/ mean 10, var.1', fontweight='bold')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('p(x)')
axs[1,0].plot(x, d_3)
axs[1,0].set_title('Student t: 2df, center 0', fontweight='bold')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('p(x)')
axs[1,1].plot(x, d_4)
axs[1,1].set_title('Student t: 2df center 10', fontweight='bold')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('p(x)')


fig.tight_layout(pad=2.0)

plt.plot()
```

```python
def posterior(p, l):
  ...
  #plots prior,likelihood, and posterior

  pos= p * l #posterior
  vals=np.array([p,l,pos])

  #now normalize each

  pnorm=(p-p.min())/(p.max()-p.min())
  lnorm=(l-l.min())/(l.max()-l.min())
  posnorm=(pos-pos.min())/(pos.max()-pos.min())


  #prior = p
  fig1=plt.figure()
  plt.plot(pnorm, label='prior')
  plt.xlabel("probability")
  plt.legend()

  #likelihood = l
  fig2=plt.figure()
  plt.plot(lnorm, label="likelihood")
  plt.xlabel("probability")
  plt.legend()

  #posterior = pos
  fig3=plt.figure()
  plt.plot(posnorm, label='posterior')
  plt.xlabel("probability")
  plt.legend()

  return fig1, fig2, fig3
```
